# 主要代码

```python
import pandas as pd
import jieba
pd.set_option('display.max_columns',None)
```

```python
date = pd.read_csv('../nft_competition/data/Assets.csv')
date.head()
```

```python
rowdate = date['description'] #获取文本列
rowdate = rowdate.dropna(axis=0, how='any') # 删除缺失值
```

```python
from collections import Counter
from multistop import Stopwords
#默认选取的中文lang='chinese'
sw = Stopwords()
sw.setlang('english')
word_count = Counter()   # 定义统计词典
```

```python
from tqdm import tqdm_notebook as tqdm
for i in tqdm(rowdate):
    # print(i.split(' '))
    for j in i.split(" "):
        if not  sw.contains(j): # 去除停用词
            word_count[j] +=1
```

## 查看top10

```python
print(word_count.most_common(10))
```

```python
out_dict = dict(word_count) # Counter 转字典
df2=pd.DataFrame.from_dict(out_dict, orient='index',columns=['value'])
df2.index.name = "word" # 修改index 名字
print(df2)
```
