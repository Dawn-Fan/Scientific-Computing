```python
import pandas as pd
import jieba
pd.set_option('display.max_columns',None)
```

```python
date = pd.read_csv('../nft_competition/data/Assets.csv')
date.head()
```

```python
rowdate = date['description'] #Ëé∑ÂèñÊñáÊú¨Âàó
rowdate = rowdate.dropna(axis=0, how='any') # Âà†Èô§Áº∫Â§±ÂÄº
```

```python
from collections import Counter
from multistop import Stopwords
#ÈªòËÆ§ÈÄâÂèñÁöÑ‰∏≠Êñálang='chinese'
sw = Stopwords()
sw.setlang('english')
word_count = Counter()   # ÂÆö‰πâÁªüËÆ°ËØçÂÖ∏
```

```python
from tqdm import tqdm_notebook as tqdm
for i in tqdm(rowdate):
    # print(i.split(' '))
    for j in i.split(" "):
        if not  sw.contains(j): # ÂéªÈô§ÂÅúÁî®ËØç
            word_count[j] +=1
```

## Êü•Áúãtop10

```python
print(word_count.most_common(10))
```

```python
out_dict = dict(word_count) # Counter ËΩ¨Â≠óÂÖ∏
df2=pd.DataFrame.from_dict(out_dict, orient='index',columns=['value'])
df2.index.name = "word" # ‰øÆÊîπindex ÂêçÂ≠ó
print(df2)
```
